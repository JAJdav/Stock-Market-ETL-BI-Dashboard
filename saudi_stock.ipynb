```python
cols = ['open','high','low']

saudi_stock = saudi_stock.sort_values(['symbol', 'date'])

saudi_stock[cols] = (
    saudi_stock.groupby('symbol')[cols]
    .transform(lambda x: x.ffill().bfill())
)
```

```python
saudi_stock.isnull().sum()
```

```python
saudi_stock[saudi_stock['name'].isnull()]
```

```python
saudi_stock['high'] = saudi_stock[['high','open','close']].max(axis=1)
saudi_stock['low']  = saudi_stock[['low','open','close']].min(axis=1)
```

```python
sector_medians = saudi_stock.groupby('sector')['no_trades'].transform('median')

saudi_stock['no_trades'] = saudi_stock.groupby('name')['no_trades'].transform(lambda x: x.fillna(x.median()))

saudi_stock['no_trades'] = saudi_stock['no_trades'].fillna(sector_medians)
```
